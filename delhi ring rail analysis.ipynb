```python
def deagree_radians(degrees):
    return degrees*(np.pi/180)
```

```python
R = 6371
def dist(lon1,lat1,lon2,lat2):
    d_lat = deagree_radians(lat2-lat1)
    d_lon = deagree_radians(lon2-lon1)
    a =  np.sin(d_lat/2)**2 + np.cos(deagree_radians(lat1)) * np.cos(deagree_radians(lat2)) * np.sin(d_lon/2) **2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return R * c
```
